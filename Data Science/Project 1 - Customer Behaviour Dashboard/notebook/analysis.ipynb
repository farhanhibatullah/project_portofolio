```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
```

```python
online_retail = pd.read_feather('../dataset/online_retail_with_date.feather')
online_retail.head()
```

Pertama kita perlu bagi menjadi dua group dataset berdasarkan tipe pembelian yaitu retail dan grocir (wholesaling)

```python
# retail dataset
retail_group = online_retail[online_retail['PurchaseType'] == 'Retail'].reset_index(drop=True)
retail_group.head()
```

```python
wholesaling_group = online_retail[online_retail['PurchaseType'] == 'Wholesaling'].reset_index(drop=True)
wholesaling_group.head()
```

Kemudian kita perlu kelompokkan tiap group tadi berdasarkan CustomerID

```python
customerid_country_retail = retail_group.groupby(by=['CustomerID', 'Country'], observed=True).agg({
    'InvoiceNo': 'count',
    'TotalPrice': 'sum',
}).rename(columns={
    'InvoiceNo': 'Frequency',
    'TotalPrice': 'Monetary'
}).reset_index()
customerid_country_retail.sort_values(by='Monetary', ascending=False).head()
```

```python
customerid_country_wholesaling = wholesaling_group.groupby(by=['CustomerID', 'Country'], observed=True).agg({
    'InvoiceNo': 'count',
    'TotalPrice': 'sum',
}).rename(columns={
    'InvoiceNo': 'Frequency',
    'TotalPrice': 'Monetary'
}).reset_index()
customerid_country_wholesaling.sort_values(by='Monetary', ascending=False).head()
```

```python
online_retail['YearOfTransaction'] = online_retail['InvoiceDate'].dt.year
online_retail['MonthOfTransaction'] = online_retail['InvoiceDate'].dt.month
online_retail['DayOfTransaction'] = online_retail['InvoiceDate'].dt.day
```

```python
revenue = online_retail.groupby(by=['YearOfTransaction', 'MonthOfTransaction'])[['TotalPrice']].agg({
    'TotalPrice':'sum'
}).rename(columns={
    'TotalPrice':'Revenue'
}).reset_index()

revenue['Period'] = revenue['MonthOfTransaction'].astype(str).str.zfill(2) + '/' + revenue['YearOfTransaction'].astype(str)
revenue['PeriodDate'] = pd.to_datetime(revenue['Period'])
revenue.head()
```

```python
online_retail['Week'] = online_retail['InvoiceDate'].dt.isocalendar().week
online_retail['YearMonth'] = online_retail['InvoiceDate'].dt.to_period('M').astype(str)
weekly_orders = online_retail.groupby(['YearMonth', 'Week'])[['Quantity']].sum().reset_index()
weekly_orders.head()
```
